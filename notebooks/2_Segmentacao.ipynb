### Segmentação

A análise RFM é uma técnica de segmentação do comportamento do cliente baseada em dados.

RFM significa **recência**, **frequência** e **valor monetário**.

O objetivo desta análise é segmentar os clientes com base em quando foi sua última compra, com que frequência eles compraram no passado e quanto gastaram no total.

Embora a análise RFM tenha nascido na mala direta, é uma ferramenta poderosa para as lojas de comércio eletrônico.
